```python
import os
import pandas as pd
import datetime
```

```python
# 读取股票代码文件
df_stock_codes = pd.read_csv('data/stock_codes.csv')
all_daily_data = pd.DataFrame()
# 遍历股票代码数据框中的每个股票代码
for stock_code in df_stock_codes['stock_codes']:
    
    file_type = 'daily'
    # 构造对应股票代码和daily表的文件名
    file_name = f'data/{file_type}_data/{stock_code}.csv'
    
    # 读取daily数据文件到数据框
    stock_data = pd.read_csv(file_name)

    all_daily_data = pd.concat([all_daily_data, stock_data])
    
daily_name = ['ts_code', 'open', 'high', 'low', 'close']
daily_data = pd.DataFrame()
all_daily_data['trade_date'] = pd.to_datetime(all_daily_data['trade_date'], format='%Y%m%d') # 将交易日期转换为日期对象
daily_data['datetime'] = all_daily_data['trade_date'] 
daily_data[daily_name] = all_daily_data[daily_name]
daily_data['volume'] = all_daily_data['vol']
daily_data['openinterest'] = 0



```

```python
daily_data
```

```python
start_date = '20210131'
date_format = "%Y%m%d"
start_date = datetime.datetime.strptime(start_date, date_format)
daily_data = daily_data[(daily_data['datetime'] > start_date)]
daily_data = daily_data.sort_values(by='datetime') # 按交易日期排序数据
daily_data = daily_data.dropna(subset=['close'])
# 保存数据
file_name = f'daily_price.csv'
daily_data.to_csv(file_name, index=False)
print(f'Saved daily_price data for daily_data to {file_name}')
```
