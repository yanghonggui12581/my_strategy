```python
from FactorGenerator_class import FactorGenerator  # 导入自定义的FactorGenerator类

# 读取包含股票代码的文件
df_stock_codes = pd.read_csv('data/stock_codes.csv')

# 遍历股票代码数据框中的每个股票代码
for stock_code in df_stock_codes['stock_codes']:
    
    # 使用FactorGenerator类为特定股票代码生成因子
    my_FactorGenerator = FactorGenerator(stock_code)
    
    # 生成所有相关因子（价值、增长、财务或技术因子）
    my_FactorGenerator.generate_all_factor()
```
