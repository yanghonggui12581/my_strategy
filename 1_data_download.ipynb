```python
pro = ts.pro_api()

# 获取沪深300指数的成分股列表
data = pro.index_weight(index_code='000300.SH', start_date='20100101', end_date='20231030')

# 获取成分股代码列表
stock_codes = data['con_code'].tolist()

# 将股票代码序列转换为集合并去重
stock_codes = list(set(stock_codes))

# 先将stock_codes转换为数据框
df_stock_codes = pd.DataFrame(stock_codes, columns=['stock_codes'])

# 存储为csv文件
df_stock_codes.to_csv('data/stock_codes.csv', index=False)
```

```python
# 创建daily_data文件夹
if not os.path.exists('data/daily_data'):
    os.makedirs('data/daily_data')

# 循环获取每个股票的历史行情数据
for code in stock_codes:
    # 获取股票代码为code的历史行情数据
    stock_data = pro.daily(ts_code=code, start_date='20100101', end_date='20231030')

    # 将数据保存到文件中
    file_name = f'data/daily_data/{code}.csv'
    stock_data.to_csv(file_name, index=False)
```
