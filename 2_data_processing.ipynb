# 将所有数据均转为月频数据

```python
import os
import pandas as pd
from Converter_class import Converter
# Converter_class 中构建了一个如何将数据转换为月频数据的类
```

```python
# 分别针对日频数据和季度数据进行月频的转换
daily_type = ['daily', 'limit', 'adj_factor', 'daily_basic', 'moneyflow']
monthly_type = ['financial', 'income', 'balance', 'cashflow']
```

```python
# 代码示例
code = '000725.SZ'
file_type = 'income'
file_name = f'data/income_data/{code}.csv'
stock_data = pd.read_csv(file_name)
my_Converter = Converter(stock_data, code, file_type)
my_Converter.convert_quarterly_to_monthly_data()
```

```python
from Converter_class import Converter

# convert_daily_to_monthly_data

# 读取股票代码文件
df_stock_codes = pd.read_csv('data/stock_codes.csv')

# 遍历股票代码数据框中的每个股票代码
for stock_code in df_stock_codes['stock_codes']:
    
    # 遍历需要转换的日频表列表中的每个表
    for file_type in daily_type:
        
        # 构造对应股票代码和日频表的文件名
        file_name = f'data/{file_type}_data/{stock_code}.csv'
        
        # 读取日频数据文件到数据框
        stock_data = pd.read_csv(file_name)

        # 创建Converter实例，并传入日频数据、股票代码和日频表名称
        my_Converter = Converter(stock_data, stock_code, file_type)
        
        # 调用Converter实例的convert_daily_to_monthly_data方法，将日频数据转换为月频数据
        my_Converter.convert_daily_to_monthly_data()

```

```python
from Converter_class import Converter
# convert_quarterly_to_monthly_data

# 读取股票代码文件
df_stock_codes = pd.read_csv('data/stock_codes.csv')

# 遍历股票代码数据框中的每个股票代码
for stock_code in df_stock_codes['stock_codes']:
    
    # 遍历需要转换的季度表列表中的每个表
    for file_type in monthly_type:
        
        # 构造对应股票代码和季度表的文件名
        file_name = f'data/{file_type}_data/{stock_code}.csv'
        
        # 读取季度数据文件到数据框
        stock_data = pd.read_csv(file_name)
        
        # 创建Converter实例，并传入季度数据、股票代码和季度表名称
        my_Converter = Converter(stock_data, stock_code, file_type)
        
        # 调用Converter实例的convert_quarterly_to_monthly_data方法，将季度数据转换为月频数据
        my_Converter.convert_quarterly_to_monthly_data()
```
